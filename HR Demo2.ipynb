```python
import pandas as pd


state_geo = 'MIgeodata.json'
state_unemployment = 'Active Employees Location by Area Code.csv'
state_data = pd.read_csv(state_unemployment)
```

```python
state_data['Area Codes'] = state_data['Area Codes'].astype(str)
```

```python
len(state_data)
```

```python
state_data = state_data.dropna()
```

```python
len(state_data)
```

```python
state_data['Area Codes'] = state_data['Area Codes'].str[:5]
```

```python
m = folium.Map(location=[42.1625408, -83.4029885 ], zoom_start=9)

chloropleth = folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=state_data,
    columns=['Area Codes', 'Active'],
    key_on='feature.properties.ZCTA',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Currently Active(%)'
).add_to(m)

folium.LayerControl().add_to(m)

```

```python
chloropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['ZCTA'])
)
```
